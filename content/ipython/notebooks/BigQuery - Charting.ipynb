For our first example we will look at some log data of Uber rides in San Francisco, and draw a trip on a map. First we should look at the data which we can do with a paged_table chart magic (there is also a table chart magic which will show the entire table; use that with caution with tables you know to be small only):

```python
%chart paged_table data-studio-team:uberlogs.logs
```

All the chart magics take a data source argument, which can be a table name, or a reference to a table, query or view.

Let's explore the data by displaying the path of a trip on a map. To do this we will use the DataLab %chart cell magic. This magic uses Google Charts to easily chart data without having to write code. The first argument is the type of chart we want, and as you can see from the help below, numerous types are supported:

```python
%chart --help
```

Following the chart type, we can specify the data to be charted. However, Google Charts requires columns in data to be in a specific order so we are often going to need to specify these as well, unless our data happens to be in the order we want. We can do this with the --field parameter which takes a comma-separated list of fields. Finally, the body of the cell can include additional options to be passed to the chart; these are chart-specific. For our map we will specify the showTip option which will show the 3rd column (the timestamp) when hovering over a map pushpin.

You can experiment with viewing different trips by changing the WHERE clause below and re-executing the chart magic.

Note in the example below we could change our SELECT statement to reorder the fields and avoid the --field argument to the chart magic but we use a different order for illustrative purposes.

```python
%%bigquery sql --name ride1
SELECT timestamp, lat, lon FROM [data-studio-team:uberlogs.logs] WHERE trip = 100 ORDER BY timestamp
```

```python
%%chart map --field lat,lon,timestamp ride1
{
  showTip: true
}
```

For our next example, lets look at the popularity of programming languages. We can use the public github_timeline data for this. We will get a rough view by counting the incidences of pushes bucketed by the repository_language field. Unlike in our other notebook where we explored github data, we won't limit the set of language in the query, but will instead use the sliceVisibilityThreshold option to the pie chart to bucket all languages with less than 6 degree pie slices (i.e. less than 1/60th of the pushes) into an 'Other' bucket.

```python
%%bigquery sql --name github_events
SELECT repository_language, COUNT(repository_language) as pushes
FROM [publicdata:samples.github_timeline]
WHERE type = 'PushEvent'
  AND repository_language != ''
GROUP BY repository_language
ORDER BY pushes DESC
```

```python
%%chart pie --field repository_language,pushes github_events
{
  title: 'Language Popularity',
  sliceVisibilityThreshold: 6/360
}
```

For a third example, let's look at weather data. The gsod sample data has weather measurements from many stations around the world. We will look at weather station 471270, which is in South Korea, from 2000 to 2010. In this example we will show that we can chart Python data, not just BigQuery tables and query results. For this station only max_temperature was measured, not min_temperature.

```python
%%bigquery sql --name skweather
SELECT year, month, day, max_temperature
FROM [publicdata:samples.gsod]
WHERE station_number = 471270 AND year > 2000
```

We need to change the separate day, month and year fields into a single datetime
to plot the temperature over time. To do this we will read the data into a Python data structure and then use that for the chart. We will use an annotation chart, which is useful for time series:

```python
import time
import datetime
weather = [{'date': datetime.datetime(year=row['year'], month=row['month'], day=row['day']), 
            'max': row['max_temperature']} for row in skweather.results()]
```

```python
%chart annotation --field date,max weather
```

For our last example, let's look at a scatter plot. This time we will look at the public natality data and plot a sample of the data, showing gestation weeks vs birth weight. We'll add some axes titles, a title and a trend line via chart options.

```python
%%bigquery sql --name babies
SELECT gestation_weeks, weight_pounds FROM [publicdata:samples.natality] WHERE gestation_weeks < 99 LIMIT 1000
```

```python
%%chart scatter babies
{
    title: 'Birth Weight vs Gestation Weeks',
    hAxis: {title: 'Gestation Weeks'},
    vAxis: {title: 'Weight'},
    legend: 'none',
    trendlines: { 0: {} }    // Draw a trendline for data series 0.
}
```
