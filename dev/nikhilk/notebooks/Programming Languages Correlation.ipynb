## Querying the Data

Notice the integration of SQL into the notebook - its just SQL in a first-class form. We'll be adding colorization, syntax checking in the future.

Queries do not execute when defined, but rather when `results()` is explicitly called. This allows composing complex queries more easily.

Here we're finding the top-25 popular languages, based on number of pushes over the last year.

### Iterative exploration and query refining

```python
%%bigquery
SELECT repository_name, type, created_at, repository_language
FROM [githubarchive:github.timeline]
LIMIT 5
```

```python
%%bigquery
SELECT repository_name, type, created_at, repository_language
FROM [githubarchive:github.timeline]
WHERE type = 'CreateEvent'
LIMIT 5
```

```python
%%bigquery
SELECT repository_name, type, created_at, repository_language
FROM [githubarchive:github.timeline]
WHERE type = 'CreateEvent'
ORDER BY created_at DESC
LIMIT 5
```

### Finding the popular languages

```python
%%bigquery popular_languages
SELECT repository_language, COUNT(repository_language) as pushes
FROM [githubarchive:github.timeline]
WHERE type = 'PushEvent'
  AND repository_language != ''
  AND PARSE_UTC_USEC(created_at) >= DATE_ADD(CURRENT_TIMESTAMP(), -1, 'YEAR')
GROUP BY repository_language
ORDER BY pushes DESC
LIMIT 25
```

Lets execute the query. Additionally, lets convert this to a standard python pandas dataframe. So BigQuery easily integrates into the python data analysis story naturally, and out-of-the-box.

```python
popular_languages.results().to_dataframe()
```

Lets do a join query - notice the use of `$popular_languages` - any python variable can be substituted into the SQL. In this case the variable is a query, so its SQL representation gets used.

This query will return pushes per user per language, again over the course of the last year, for just the popular languages.

Again, we're just defining the query, not yet executing it.

### Finding the push events
Only for the popular languages... otherwise there are too many languages just adding noise.

```python
%%bigquery pushes
SELECT timeline.actor AS user, timeline.repository_language AS language, COUNT(timeline.repository_language) AS pushes
FROM [githubarchive:github.timeline] AS timeline
JOIN $popular_languages AS languages
  ON timeline.repository_language = languages.repository_language
WHERE type = 'PushEvent'
  AND PARSE_UTC_USEC(created_at) >=
      DATE_ADD(CURRENT_TIMESTAMP(), -1, 'YEAR')
GROUP BY user, language
```

Lets nest that query into another SQL query that adds hashing to perform a sampling ... we'll take a 5% sample.

```python
%%bigquery sampled_pushes
SELECT user, language, pushes
FROM $pushes
WHERE ABS(HASH(user)) % 100 < 5
```

And then execute the query, again converting the result-set into a dataframe. Lets also see the length, and a few rows of data to get a sense of what we have to work with.

```python
df = sampled_pushes.results().to_dataframe()
len(df)
```

## Data Analysis in Python

The results we have so far, have multiple rows for each user. One per language. Pivot that data, so we have one row per user, and one column per language.

Pandas offers a built-in `pivot()` method.

```python
dfp = df.pivot(index = 'user', columns = 'language', values = 'pushes') \
        .fillna(0)
dfp
```

And then compute the correlation for each pair of columns, again, built-in into the pandas library.

```python
corr = dfp.corr(method = 'spearman')
corr
```

## Data Visualization

All those numbers are great, but seeing a plot will make the data speak a lot louder. Red for least correlated, and blue for the most.

The visualization is done using matplotlib, which is the defacto library used in ipython notebooks.

**Future note**: we're going to look into building a more purpose/scenario-driven visualization library based on d3.js (think interactive visualization) that is language agnostic.

```python
def plot_correlation(data, title = '', type = ''):
    min_value = 0
    max_value = 0

    for i in range(len(data.columns)):
        for j in range(len(data.columns)):
            if i != j:
                min_value = min(min_value, data.iloc[i, j])
                max_value = max(max_value, data.iloc[i, j])
    span = max(abs(min_value), abs(max_value))
    span = round(span + .05, 1)

    items = data.columns.tolist()
    ticks = np.arange(0.5, len(items) + 0.5)

    plot.figure(figsize = (11, 7))
    plot.pcolor(data.values, cmap = 'RdBu', vmin = -span, vmax = span)
    plot.colorbar().set_label(type)
    plot.title(title)
    plot.xticks(ticks, items, rotation = 'vertical')
    plot.yticks(ticks, items)
    plot.show()
```

```python
plot_correlation(corr, 'Language Correlation', 'Spearman Correlation')
```
