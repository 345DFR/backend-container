# Custom Metrics from Google Analytics Data

```python
import gcp.interactive
import gcp.bigquery as bq
```

# Data

This notebook works against a BigQuery dataset that represents a sample of Google Analytics data (visits, i.e. sessions containing individual hit events).

```python
event_log_table = bq.table('google.com:analytics-bigquery:LondonCycleHelmet.ga_sessions_20130910')
```

### Exploring the Data

A simple query to sample the data - seeing visits and hits within a visit flattened into a table.

```python
%%bigquery visits
SELECT visitId, visitStartTime, hits.page.pagePath AS path, hits.hitNumber AS hitNumber, totals.Hits as totalHits
FROM $event_log_table
ORDER BY visitId, hitNumber
```

```python
visits.sample(15).to_dataframe()
```

### Filtering the Data

Only the visits that included a hit to the login page.

```python
%%bigquery visits_with_logins
SELECT visitId, visitStartTime, hits.page.pagePath AS path, hits.hitNumber AS hitNumber, totals.hits as totalHits,
       SUM(IF(hits.page.pagePath == '/login.html', 1, 0)) WITHIN RECORD AS logins
FROM $event_log_table
HAVING logins > 0
ORDER BY visitId, hitNumber
```

```python
visits_with_logins.sample(10).to_dataframe()
```

### Flattening to just the Login Hits

Build a nested query to get just hits to the login endpoint, and also flag those hits as dropped if the hit to the login page is the last hit within the visit.

```python
%%bigquery login_hits
select visitId, visitStartTime, hitNumber, totalHits,
       IF(hitNumber == totalHits, 1, 0) as bounced
from $visits_with_logins
WHERE path == '/login.html'
```

```python
login_hits.sample(10).to_dataframe()
```

### Aggregating

First aggregate to counts, and then to a single metric.

```python
%%bigquery activity
SELECT SUM(IF(bounced == 0, 1, 0)) AS complete,
       SUM(IF(bounced == 1, 1, 0)) AS incomplete
FROM $login_hits
```

```python
activity.results().to_dataframe()
```

```python
%%bigquery metric
SELECT incomplete * 100 / (complete + incomplete) AS failureRate
FROM $activity
```

```python
metric.results().to_dataframe()
```

### Time Series

Continue aggregating, but first group over a fixed interval to create a time-series instead.

```python
%%bigquery activity_timeseries
SELECT SUM(IF(bounced == 0, 1, 0)) AS complete,
       SUM(IF(bounced == 1, 1, 0)) AS incomplete,
       (visitStartTime / 300000) * 300000 AS window
FROM $login_hits
GROUP BY window
```

```python
activity_timeseries.results().to_dataframe()
```

```python
%%bigquery metric_timeseries
SELECT incomplete * 100 / (complete + incomplete) AS failureRate
FROM $activity_timeseries
```

```python
df = metric_timeseries.results().to_dataframe()
df
```

```python
figsize(11, 7)

datapoints = len(df)
ticks = array(range(datapoints))

plot(ticks, df['failureRate'])
grid()
axis(ymin=0, ymax=100)

plot()
```
